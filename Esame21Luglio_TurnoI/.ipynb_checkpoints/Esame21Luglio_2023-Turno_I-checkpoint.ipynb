```python
from scipy.io import loadmat

import numpy as np

dati = loadmat('testI.mat')

A=dati["A"] 

A=A.astype(float)

b=dati["b"] 

b=b.astype(float)
```

```python
def checkIfDescentMethodCanBeApply(A):
    print(A.shape)
    if min(A.shape[0],A.shape[1])>=200:
        print("grandi dimensioni")
        non_zero = np.count_nonzero(A)
        perc = non_zero/(A.shape[0]*A.shape[1]) * 100
        print(perc,"%")
        if perc <= 33:
            print("matrice sparsa")
            if np.all(A.T == A):
                print("simmetrica")
                autoVal,autoVet = np.linalg.eigh(A)
                flag = autoVal>0
                if np.all(flag):
                    print("def pos")
                    print("possono essere usati")

checkIfDescentMethodCanBeApply(A)
```

```python
def steepestdescent(A,b,x0,itmax,tol):
 
    n,m=A.shape
    if n!=m:
        print("Matrice non quadrata")
        return [],[]
    
    
   # inizializzare le variabili necessarie
    x = x0

     
    r = A@x-b
    p = -r
    it = 1
    nb=np.linalg.norm(b)
    errore=np.linalg.norm(r)/nb
    vec_sol=[]
    vec_sol.append(x)
    vet_r=[]
    vet_r.append(errore)
     
# utilizzare il metodo del gradiente per trovare la soluzione
    while it<=itmax and errore>=tol:
        it=it+1
        Ap= A@p
       
        alpha = -np.dot(r.T,p)/np.dot(p.T,Ap)
                
        x = x+x*alpha
        
         
        vec_sol.append(x)
        r=r+alpha*Ap
        errore=np.linalg.norm(r)/nb
        vet_r.append(errore)
        p =-r
        
     
    return x,vet_r,vec_sol,it


def conjugate_gradient(A,b,x0,itmax,tol):
    n,m=A.shape
    if n!=m:
        print("Matrice non quadrata")
        return [],[]
    
    
   # inizializzare le variabili necessarie
    x = x0
    
    r = A@x-b
    p = -r
    it = 1
    nb=np.linalg.norm(b)
    errore=np.linalg.norm(r)/nb
    vec_sol=[]
    vec_sol.append(x0)
    vet_r=[]
    vet_r.append(errore)
# utilizzare il metodo del gradiente coniugato per calcolare la soluzione
    while errore >= tol and it< itmax:
        it=it+1
        Ap=A@p
        alpha = -(np.dot(r.T,p)/np.dot(p.T,Ap))
        x =x+x*alpha
        vec_sol.append(x)
        rtr_old=r.T@r
        r=r+alpha*Ap
        gamma= np.dot(r.T,r)/rtr_old
        errore=np.linalg.norm(r)/nb
        vet_r.append(errore)
        p = -r+gamma*p
   
    
    return x,vet_r,vec_sol,it
```

```python
import matplotlib.pyplot as plt

tol = 1e-6
itmax = 2000

x0 = np.zeros((A.shape[0],1))

xS,vetS,solS,itS = steepestdescent(A,b,x0,itmax,tol)

xG,vetG,solG,itG = conjugate_gradient(A,b,x0,itmax,tol)
print("steepestdescent it => ",itS)
print("conjugate_gradient it => ",itG)

plt.semilogy(range(itS),vetS)
plt.semilogy(range(itG),vetG)

plt.legend(["steepestdescent","conjugate_gradient"])
plt.xlabel("iterations")
plt.ylabel("error")
plt.show()
```

la matrice ha indice di condizionamento pari a $$10^4$$ quindi è mal condizionata, questo implica che l'errore si propaga di molto sulla soluzione ad ogni iterazione fatta.

```python
A1 = A.copy()
b1=np.ones(b.shape)
diag= np.zeros(A.shape)
for i in range(A.shape[0]):
    diag[i][i] = 0.05
A1 = A1+diag

xS,vetS,solS,itS = steepestdescent(A1,b1,x0,itmax,tol)

xG,vetG,solG,itG = conjugate_gradient(A1,b1,x0,itmax,tol)
print("steepestdescent it => ",itS)
print("conjugate_gradient it => ",itG)

plt.semilogy(range(itS),vetS)
plt.semilogy(range(itG),vetG)

plt.legend(["steepestdescent","conjugate_gradient"])
plt.xlabel("iterations")
plt.ylabel("error")
plt.show()

print("condizionamento di A1 => {:e}".format(np.linalg.cond(A1)))
```

Scrivere lo script  in cui
 - si determini il polinomio di approssimazione ai minimi quadrati di grado 1 dei  punti assegnati sviluppando  le functions necessarie; **Punti: 4**
- si determini il polinomio di interpolazione dei punti assegnati sviluppando le functions necessarie;  **Punti: 4**
- si consideri il set di dati ($x_i,\hat{y}_i$) , dove $\hat{y}_i = 0.2 x_i+1$, $i = 0, ..., 10$ e si ricalcolino il polinomio di approssimazione ai minimi quadrati di grado 1 e il polinomio di interpolazione a partire dai nuovi punti $(x_i, \hat{y}_i)$, i = 0, ..., 10; **Punti: 1**
- si rappresentino in figura 1 (rispettivamente in figura 2) il primo (rispettivamente il secondo) set di punti assegnato insieme al polinomio di approssimazione ai minimi quadrati e al polinomio di interpolazione ottenuti da questi dati; **Punti: 1**

**Totale:  14**

```python
import scipy.linalg as spLin
import RisolviSis as SolveTriangular
def qrLS(A,b):
#Risolve un sistema sovradeterminato con il metodo QR-LS
    n=A.shape[1]  # numero di colonne di A
    Q,R=spLin.qr(A)
    h=Q.T@b
    x,flag=SolveTriangular.Usolve(R[0:n,:],h[0:n])
    residuo=np.linalg.norm(h[n:])**2
    return x,residuo

def plagr(xnodi,j):
    """
    Restituisce i coefficienti del j-esimo pol di
    Lagrange associato ai punti del vettore xnodi
    """
    xzeri=np.zeros_like(xnodi)
    n=xnodi.size
    if j==0:
       xzeri=xnodi[1:n]
    else:
       xzeri=np.append(xnodi[0:j],xnodi[j+1:n])
    
    num=np.poly(xzeri)
    den=np.polyval(num,xnodi[j])
    
    p=num/den
    
    return p

def InterpL(x, y, xx):
     """"
        %funzione che determina in un insieme di punti il valore del polinomio
        %interpolante ottenuto dalla formula di Lagrange.
        % DATI INPUT
        %  x  vettore con i nodi dell'interpolazione
        %  f  vettore con i valori dei nodi 
        %  xx vettore con i punti in cui si vuole calcolare il polinomio
        % DATI OUTPUT
        %  y vettore contenente i valori assunti dal polinomio interpolante
        %
     """
     n=x.size
     m=xx.size
     L=np.zeros((m,n))
     for j in range(n):
        p=plagr(x,j)
        L[:,j]=np.polyval(p,xx)
    
    
     return L@y
```

```python
x = np.array([1.0, 1.2, 1.4, 1.6, 1.8, 2.0, 2.2, 2.4, 2.6, 2.8, 3.0]) 

y = np.array([1.18, 1.26, 1.23, 1.37, 1.37, 1.45, 1.42, 1.46, 1.53, 1.59, 1.59])

plt.plot(x,y,"*")

n=1
A1 = np.vander(x,increasing=True)[:,:n+1]
x1,residuo = qrLS(A1,y)


xx = np.linspace(np.min(x),np.max(x),200)
y1 = np.polyval(np.flip(x1),xx)
plt.plot(xx,y1)
plt.show()
```

```python
interpolator = InterpL(x,y,xx)

plt.plot(x,y,"*")
plt.plot(xx,interpolator)
plt.show()
```

```python
x1 = x.copy()
y1 = 0.2*x1 +1

plt.plot(x1,y1,"*")

xx1 = np.linspace(min(x),max(x),200)

n = 1 
A1 = np.vander(x1,increasing=True)[:,:n+1]
x_1,res=qrLS(A1,y1)
pol = np.polyval(np.flip(x_1),xx1)
plt.plot(xx1,pol)
plt.show()
plt.plot(x1,y1,"*")

inter=InterpL(x1,y1,xx1)
plt.plot(xx,inter)
```

Noto che l'approssimazione è perfetta, quindi E(x) = 0.
Questo accade perchè x_perturbata è uguale a x data dal problema, secondo il teorema dell'errore:
$$E(x) = f(x)-P_n(x)$$
quindi E(x) = 0 quando x è uguale, oppure se la derivata w(n+1) si annulla
